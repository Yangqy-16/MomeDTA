```python
import pandas as pd
import matplotlib.pyplot as plt

# 
excel_file = pd.ExcelFile('./ablation/ablation.xlsx')

# 
df = excel_file.parse('Sheet1')

# 
plt.rcParams['figure.dpi'] = 300

# 
# plt.rcParams['font.sans-serif'] = ['WenQuanYi Zen Hei']

#  Method 
df.set_index('Methods', inplace=True)

# 
df_transposed = df.T

#  - 
colors = ['#7b5da0', '#427AB2', '#ABD0F1', '#43978F', '#E56F5E', '#F6C957', '#FFB77F']
#7b5da0#9EC4BE
# 
plt.figure(figsize=(10, 6))

# 
ax = df_transposed.plot(
    kind='bar', 
    color=colors,
    width=0.8,  # 
    edgecolor='black',  # 
    linewidth=0.8  # 
)

# 
legend = ax.legend(fontsize=11, loc='best', frameon=True, framealpha=0.9)

#  X 
plt.xticks(rotation=360)

# 
ax.grid(True, axis='y', linestyle='--', alpha=0.7)

# 
ax.tick_params(axis='both', which='major', labelsize=14)

# 
plt.tight_layout()

# 
plt.savefig('./ablation/ablation.png', dpi=200, bbox_inches='tight')

# 
plt.show()
```
